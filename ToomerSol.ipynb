```python
import numpy as np
import matplotlib.pyplot as plt

from numba import jit
from typing import List,Dict,Any
```

```python
class N_System:
    def __init__(self, particles: List[Dict[str, Any]]):
        ''' Clase general para un sistema de N partículas, donde cada partícula es un diccionario con sus propiedades
        - mass: masa de la partícula
        - position: posición inicial (array de 3D)
        - velocity: velocidad inicial (array de 3D)'''
        self.n = len(particles)
        self.positions = np.array([p['position'] for p in particles])
        self.velocities = np.array([p['velocity'] for p in particles])
        self.masses = np.array([p['mass'] for p in particles])
        self.systemmass = self.masses.sum()

        v_prom = np.mean(np.linalg.norm(self.velocities, axis=1))
        self.t_relax = self.n/np.log(self.n)*(v_prom**3) \
            / (self.mass) # Tiempo de relajación estimado

       
        self.mass_indices = np.where(self.masses > 0)[0]

    def evolucionar(self, dt: float, T:float
                )-> np.ndarray[float,3]:
                for _ in range(steps):
    def acc(pos: np.ndarray[float,3],
    masses:np.ndarray[float],
    m_indices: np.ndarray[int],
    n_particles:int
    ) -> np.ndarray[float,3]:

    '''Calcula la aceleración gravitatoria de cada partícula'''
        acc = np.zeros_like(n_particles, pos)
        for i in range(self.n):
            for j in m_indices:
                if i==j:
                    continue
            r_vec = pos[i] - pos[j] # Vector de posición relativa
            r_mag = np.linalg.norm(r_vec) +1e-10 # Evitar división por cero
            acc[i] -= masses[j]*r_vec/r_mag**3 # Ley de gravitación universal
        return acc
      num_steps = int(T/dt)
      trajectories = np.zeros((num_steps, self.n, 3))

      #leap-frog integration
      accelerations = self.acc(self.positions, self.masses, self.mass_indices, self.n)
      self.velocities += 0.5*accelerations*dt # Kick inicial
        for step in range(num_steps):
            self.positions += self.velocities*dt # Drift
            accelerations = self.acc(self.positions, self.masses, self.mass_indices, self.n) # Recalcular aceleraciones
            self.velocities += accelerations*dt # Kick
            trajectories[step] = self.positions # Guardar posiciones para visualización
        return trajectories
```
